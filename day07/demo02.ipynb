### 第一题

```python
import matplotlib.pyplot as plt
import numpy as np
data = np.load("./tmp/国民经济核算季度数据.npz", allow_pickle = True)
dataFile = data.files
data1 = data["columns"]
data2 = data["values"]
print(dataFile)
print(data1)
print(data2)
```

### 第二题

```python
#解决中文显示问题
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
plt.title("2000年第一季度各产业国民生产总值分布直方图")
num1 = data2[0, 3: 6]
plt.bar(range(len(num1)), num1)
plt.xticks(range(len(num1)), ["第一季度", "第二季度", "第三季度"])
plt.show()
```

### 第三题

```python
plt.title("2000年第一季度各产业国民生产总值分布饼图")
plt.pie(num1, autopct="%0.2f %%", labels=["第一产业", "第二产业", "第三产业"], labeldistance= 1.1)
plt.show()
```

### 第四题

```python
plt.title("2000-2017年各产业国民生产总值箱线图")
num2 = data2[:, 3: 6]
plt.boxplot(num2)
plt.show()
```
