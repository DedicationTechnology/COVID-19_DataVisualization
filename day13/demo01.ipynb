```python
# 散点图的绘制
from pyecharts.charts import Scatter  # 使用pyecharts绘制散点图要导入的模块
import pyecharts.options as opts  # 该模块用于图标的一些全局配置
import numpy as np
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
"""以下为绘制散点图的规则"""
(Scatter()  # 表示绘制散点图
.add_xaxis(xaxis_data = x)  # 用于添加x轴的数据
# """series_name必须写，否则可能会报错，可以写成series_name=""的形式，render_notebook()用于直接在前端显示绘图的结果"""
# label_opts=opts.LabelOpts(is_show=False)表示不显示每个点的y轴所对应的数据
.add_yaxis(series_name = "sin", y_axis = y1, label_opts=opts.LabelOpts(is_show=False))  # 用于添加y轴的数据
.add_yaxis(series_name = "cos", y_axis = y2, label_opts=opts.LabelOpts(is_show=False))).render_notebook()
```

```python
# 线图的绘制
from pyecharts.charts import Line
import pyecharts.options as opts
import numpy as np
x = np.linspace(0, 2 * np.pi, 100)
y1 = np.sin(x)
y2 = np.cos(x)
(
    Line()
    .add_xaxis(xaxis_data=x)
    .add_yaxis(series_name="sin", y_axis=y1, label_opts = opts.LabelOpts(is_show=False))
    .add_yaxis(series_name="cos", y_axis=y2, label_opts = opts.LabelOpts(is_show=False))
    # title_opts=opts.TitleOpts(title= "曲线")用于给图标设置标题   
    # tooltip_opts=opts.TooltipOpts(axis_pointer_type="cross")用于显示交互设置中进行十字形坐标显示，点击某个点会显示横坐标和纵坐标
    .set_global_opts(title_opts=opts.TitleOpts(title= "曲线"),
                    tooltip_opts=opts.TooltipOpts(axis_pointer_type="cross"))  
).render()  # render()可以在指定的目录下生成一个html文件用于显示图标的结果
```

```python
 # 饼图的绘制
from pyecharts.charts import Pie
import pyecharts.options as opts

num = [110, 136, 108, 48, 111, 112, 103]
lab = ['哈士奇', '萨摩耶', '泰迪', '金毛', '牧羊犬', '吉娃娃', '柯基']

(
    Pie(init_opts=opts.InitOpts(width='720px', height='320px'))  # init_opts=opts.InitOpts(width='720px', height='320px'用于设置画布的大小
    .add(series_name='', 
         # 注意这里饼图的数据包括两个部分，分别为图例和具体的数值，必须先是图例然后是数值，顺序不能够颠倒
         data_pair=[(j, i) for i, j in zip(num, lab)]  # data_pair=[(j, i) for i, j in zip(num, lab)]用于饼图的相关数据的设置
        )
).render_notebook()
```

```python
# 环图的绘制
from pyecharts.charts import Pie
import pyecharts.options as opts

num = [110, 136, 108, 48, 111, 112, 103]
lab = ['哈士奇', '萨摩耶', '泰迪', '金毛', '牧羊犬', '吉娃娃', '柯基']

(
    Pie(init_opts=opts.InitOpts(width='720px', height='320px'))
    .add(series_name='', 
         radius=['40%', '75%'],  # 两个参数分别表示要设置饼图的内半径和外半径
         data_pair=[(j, i) for i, j in zip(num, lab)]
        )
).render_notebook()
```

```python
# 玫瑰图的绘制
from pyecharts.charts import Pie
import pyecharts.options as opts

num = [110, 136, 108, 48, 111, 112, 103]
lab = ['哈士奇', '萨摩耶', '泰迪', '金毛', '牧羊犬', '吉娃娃', '柯基']

(
    Pie(init_opts=opts.InitOpts(width='720px', height='320px'))
    .add(series_name='', 
         rosetype='radius',  # 表示玫瑰图展示，用半径大小来表示数值的大小
         data_pair=[(j, i) for i, j in zip(num, lab)]
        )
).render_notebook()
```

```python
# 柱图的绘制
from pyecharts.charts import Bar
import pyecharts.options as opts

num = [110, 136, 108, 48, 111, 112, 103]
num2 = [90, 110, 101, 70, 90, 120, 99]
lab = ['哈士奇', '萨摩耶', '泰迪', '金毛', '牧羊犬', '吉娃娃', '柯基']

(
    Bar(init_opts=opts.InitOpts(width='720px', height='320px'))
    .add_xaxis(xaxis_data=lab)
    .add_yaxis(series_name='商家A', y_axis=num)
    .add_yaxis(series_name='商家B', y_axis=num2)
    .set_global_opts(
        title_opts=opts.TitleOpts(title='各商家拥有犬类数量情况', subtitle='如有雷同，纯属意外')  # title表示主标题，subtitle表示副标题
    )
).render_notebook()
```

```python
from pyecharts.charts import Bar, Line

num = [110, 136, 108, 48, 111, 112, 103]
lab = ['哈士奇', '萨摩耶', '泰迪', '金毛', '牧羊犬', '吉娃娃', '柯基']

bar = (
    Bar(init_opts=opts.InitOpts(width='720px', height='320px'))
    .add_xaxis(xaxis_data=lab)
    .add_yaxis(series_name='', y_axis=num)
)

lines = (
    Line()
    .add_xaxis(xaxis_data=lab)
    .add_yaxis(series_name='', y_axis=num, label_opts=opts.LabelOpts(is_show=False))
)
# overlap用于添加图形，这里为在柱状图上添加折线图
bar.overlap(lines).render_notebook()
```
