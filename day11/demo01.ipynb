```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 2 * np.pi, 100)  # linspace函数用于创建等差数列
y = np.sin(x) + np.random.rand(100)  # np模块下的random模块下的rand函数用于创建0~1之间的随机数，100是随机数的个数
plt.figure(figsize=(5, 5))
plt.rcParams["font.sans-serif"] = "SimHei"  # 修改字体的样式可以解决标题中文显示乱码的问题
plt.rcParams["axes.unicode_minus"] = False  # 该项可以解决绘图中的坐标轴负数无法显示的问题
plt.title("sin散点图")
plt.scatter(x, y)
plt.show()
```

```python
x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x) + np.random.rand(100)
plt.figure(dpi=140)  # dpi表示像素的值，直观的感受就是图片大小的变化
plt.plot(x, y, "r--")  # r表示颜色为红色，--表示将实现变为虚线
plt.plot(x, np.sin(x), "g")
plt.plot([1, 2, 3, 4], "b")  # 只输入一个列表，系统默认为y的值，然后从零开始自动生成x的值，x的长度和y的长度相同，也就是说x为[0, 1, 2, 3]
plt.ylabel("y轴")  # 表示y的标签
plt.show()
```

```python
x = range(10)  # 柱状图中的x表示每一个柱形的横坐标的中心值，range(10)函数生成一个从0到9的整型数
np.random.seed(123)  # seed()可以保证只要后面的数字相同，每次生成的随机数都是一样的
y1 = np.random.random(10)
y2 = np.random.random(10)
plt.title("$\pi$")  # $$之间可以插入一些特殊的TeX方程表达式，这里可以输出Π
plt.bar(x, y1, facecolor="r")  # facecolor表示柱形的颜色
plt.bar(x, -y2, facecolor="g")
"""text用于给图添加注释，i和j分别表示要添加注释的横坐标和纵坐标，"%.2f" %j是要添加的内容，这里为保留两位小数，ha表示注释横向的位置(相对于每一个柱而言)，
va表示纵向的位置，bottom表示浮动在图的上方"""
for i, j in zip(x, y1):
    plt.text(i, j, "%.2f" %j, ha="center", va="bottom")
for i, j in zip(x, y2):
    plt.text(i, -j, "%.2f" %j, ha="center", va="top")
plt.show()
```

```python
a = np.ones(10)  # 生成10个全1的数组
"""autopct用户显示值，%i表示显示整数,%%表示显示%号,explode表示是否向外突出显示，0.1表示第一份向外突出0.1， [0] * 9表示其余9份不突出，
labels表示要添加的标签，labeldistance表示标签的突出距离"""
plt.pie(a, autopct="%0.2f%%", explode=[0.1] + [0] * 9, labels="1234567890", labeldistance=1.1)
plt.show()
```
