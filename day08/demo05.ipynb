```python
import pandas as pd
import numpy as np
from sqlalchemy import create_engine
engin = create_engine("mysql+pymysql://root:123456@127.0.0.1:3306/test?charset=utf8mb4")
data = pd.read_sql("meal_order_detail1", con=engin)
# pivot_table方法可以实现透视表,aggfunc用于接收函数,index表示按什么进行分组
# pd.pivot_table(data[["order_id", "counts", "amounts"]], index="order_id", aggfunc=np.sum)
# index中的内容的顺序直接决定了排序的顺序，这里如果将order_id写在dishes_name后面就会先按照order_id进行分组
"""pd.pivot_table(data[["order_id", "counts", "amounts", "dishes_name"]], index=["order_id", "dishes_name"], 
aggfunc=np.sum)"""
# index表示行名称，columns表示列名称，margins表示求和，会出现一个名为ALL的行
pd.pivot_table(data[["order_id", "counts", "amounts", "dishes_name"]], index="order_id", 
               columns="dishes_name", margins=True, aggfunc=np.sum)
```

```python
# 交叉表的创建，主要计算分组频率,values表示要求和的属性,fillna(0)可以将NaN强制转化为0
pd.crosstab(index=data["order_id"], columns=data["dishes_name"], values=data["counts"], margins=True, 
            aggfunc=np.sum).fillna(0)
```
