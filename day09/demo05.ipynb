```python
# 招聘数据探索分析
import pandas as pd
import re
job_info = pd.read_csv("./tmp/job_info.csv", encoding="gbk")
job_info.columns = ["公司","岗位","工作地点","工资","发布日期"]
# job_info = job_info.rename(columns={"字节跳动":"公司","数据产品经理":"岗位","北京":"工作地点","2-3.5万/月":"工资",
#                     "09-03":"发布日期"})
job_info["岗位"].value_counts().idxmax()  # idxmax()可以寻找最大值的索引
# job_info.loc[job_info["发布日期"]=="09-03", :]
job_info[job_info["发布日期"]=="09-03"]
a = job_info["工作地点"].agg(lambda x : "深圳" in x)  # 返回值是bool类型
b = job_info["岗位"]=="数据分析师"  # 返回值为bool类型
job_info[a & b]  # 取ab都为bool类型的值
def get_number(string=None):
    try:
        if string[-3] == "万":
            c = [float(i) * 10000 for i in re.findall("\d+\.*\d*", string)]
        elif string[-3] == "千":
            c = [float(i) * 1000 for i in re.findall("\d+\.*\d*", string)]
        if string[-1] == "年":
            c = [i/12 for i in c]
        return c
    except:
        return None
mid = job_info["工资"].agg(get_number)
job_info["最低工资"] = mid.str[0]
job_info["最高工资"] = mid.str[1]
job_info
```
