```python
"""静态网页的爬取"""
import urllib3 # 导入爬虫所需要的库
http = urllib3.PoolManager()  # 创建实例
rq = http.request("GET",url="https://dedicationtechnology.github.io")  # GET表示请求的类型
print("服务器响应码:", rq.status)  # status可以返回服务器响应码，并因此来查看是否请求成功，200表示请求成功  
print("相应实体:",rq.data)  # 返回的内容是网页的源代码
```

```python
# 请求头的设置
import urllib3 # 导入爬虫所需要的库
http = urllib3.PoolManager()  # 创建实例
"""在request方法中，如果需要传入headers参数，可通过定义一个字典类型实现。定义一个包含User-Agent信息的字典，
使用浏览器为火狐和chrome浏览器，操作系统为“Windows NT 6.1; Win64; x64”"""
head = {"User-Agent":"Windows NT 6.1; Win64; x64"}  # 设置请求头的值
rq = http.request("GET",url="https://dedicationtechnology.github.io", headers=head)  # GET表示请求的类型,headers参数用于请求头
print("服务器响应码:", rq.status)  # status可以返回服务器响应码，并因此来查看是否请求成功，200表示请求成功  
print("相应实体:",rq.data)  # 返回的内容是网页的源代码
```

```python
# 第一种timeout设置方法
import urllib3 # 导入爬虫所需要的库
http = urllib3.PoolManager()  # 创建实例
"""在request方法中，如果需要传入headers参数，可通过定义一个字典类型实现。定义一个包含User-Agent信息的字典，
使用浏览器为火狐和chrome浏览器，操作系统为“Windows NT 6.1; Win64; x64”"""
head = {"User-Agent":"Windows NT 6.1; Win64; x64"}
# 第一种timeout设置方法
"""为防止因为网络不稳定、服务器不稳定等问题造成连接不稳定时的丢包，可以在请求中增加timeout参数设置"""
rq = http.request("GET",url="https://dedicationtechnology.github.io", headers=head, timeout= 3.0)  # timeout参数用于设置超时时间，超时就终止
print("服务器响应码:", rq.status)  # status可以返回服务器响应码，并因此来查看是否请求成功，200表示请求成功  
print("相应实体:",rq.data)  # 返回的内容是网页的源代码
```

```python
# 第二种timeout设置方法
import urllib3 # 导入爬虫所需要的库
http = urllib3.PoolManager()  # 创建实例
"""在request方法中，如果需要传入headers参数，可通过定义一个字典类型实现。定义一个包含User-Agent信息的字典，
使用浏览器为火狐和chrome浏览器，操作系统为“Windows NT 6.1; Win64; x64”"""
head = {"User-Agent":"Windows NT 6.1; Win64; x64"}
# 第二种timeout设置方法
"""为防止因为网络不稳定、服务器不稳定等问题造成连接不稳定时的丢包，可以在请求中增加timeout参数设置"""
rq = http.request("GET",url="https://dedicationtechnology.github.io", headers=head, 
                  timeout=urllib3.Timeout(connect=1.0, read=2.0))  # urllib3.Timeout(connect=1.0, read=2.0)表示连接时间超过1秒，读取时间超过2秒就停止请求
print("服务器响应码:", rq.status)  # status可以返回服务器响应码，并因此来查看是否请求成功，200表示请求成功  
print("相应实体:",rq.data)  # 返回的内容是网页的源代码
```

```python
# 第三种timeout设置方法
import urllib3 # 导入爬虫所需要的库
# 第三种timeout设置方法
http = urllib3.PoolManager(timeout=4.0)  # 创建实例,表示后续所有的操作都按这个超时时间进行
"""在request方法中，如果需要传入headers参数，可通过定义一个字典类型实现。定义一个包含User-Agent信息的字典，
使用浏览器为火狐和chrome浏览器，操作系统为“Windows NT 6.1; Win64; x64”"""
head = {"User-Agent":"Windows NT 6.1; Win64; x64"}
"""为防止因为网络不稳定、服务器不稳定等问题造成连接不稳定时的丢包，可以在请求中增加timeout参数设置"""
rq = http.request("GET",url="https://dedicationtechnology.github.io", headers=head ) 
print("服务器响应码:", rq.status)  # status可以返回服务器响应码，并因此来查看是否请求成功，200表示请求成功  
print("相应实体:",rq.data)  # 返回的内容是网页的源代码
```

```python
# 第一种请求重试设置(就是第一次没有请求成功，重新申请请求)
import urllib3
http = urllib3.PoolManager(timeout=4.0, retries=10)  # retries=10表示重新请求10次
head = {"User-Agent":"Windows NT 6.1; Win64; x64"}
rq = http.request("GET",url="https://dedicationtechnology.github.io", headers=head ) 
print("服务器响应码:", rq.status)  
print("相应实体:",rq.data)  
```

```python
# 使用urllib3进行完整的HTTP请求
import urllib3
# 发送请求实例
http = urllib3.PoolManager()  
# 请求头的具体信息到网站上用开发者工具进行复制粘贴
head = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/84.0.4147.89 Safari/537.36"}
# 要爬取的网址
url = "https://dedicationtechnology.github.io"
# 设置超时时间
tm = urllib3.Timeout(connect=1.0, read=3.0)  
# 重试次数和重定向次数
rq = http.request("GET",url=url, headers=head, timeout=tm, retries=5, redirect=4) # redirect表示重定向的次数
print("服务器响应码:", rq.status)  
# 如果不确定网站的编码格式，可以先用rq.data来获取网站的源代码，然后再顶部寻找<meta charset="utf-8">这类似的属性，会告诉你编码格式
print("相应实体:",rq.data.decode("utf8"))  # decode("utf8")可以来保证可以将中文输出正确
```

```python
# 使用request进行网页的请求
import requests
url = "https://dedicationtechnology.github.io"
rq = requests.get(url)  # get表示请求的类型
rq.encoding= "utf-8" # 设定编码方式
print("响应码:", rq.status_code)
print("编码:", rq.encoding)
print("请求头:", rq.headers)
# print("实体:", rq.text)  # 第一种访问实体的方式
print("实体:", rq.content.decode("utf-8"))  # 第二种访问实体的方式,content必须使用decode方法来进行编码的指定
```
