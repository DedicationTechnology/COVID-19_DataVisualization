#### 构建并评估回归模型

```python
 # 数据集的导入(这里导入波士顿房价数据集)
from sklearn.datasets import load_boston
data = load_boston()
x = data["data"]
y = data["target"]
```

```python
# 数据集的切分
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# 模型的构建(这里构建线性回归模型)
from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(x_train, y_train)  # LinearRegression后面必须要跟fit后才能跟训练集
y_pre = model.predict(x_test)
```

```python
# 模型的评价
import matplotlib.pyplot as plt
#解决中文显示问题
plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus'] = False
# 这里通过比较真实输出和预测输出的折线图来判断模型的优劣
for i in range(2):
    plt.plot(range(len(y_test)), y_test)
    plt.plot(range(len(y_pre)), y_pre)
plt.legend(["真实结果", "预测结果"])
plt.show()
"""mean_absolute_error, r2_score这两个函数也可以判定模型的优劣"""
from sklearn.metrics import mean_absolute_error, r2_score
mean_absolute_error(y_test, y_pre)  # mean_absolute_error结果越接近0越好
r2_score(y_test, y_pre)  # r2_score结果越接近1越好
```
