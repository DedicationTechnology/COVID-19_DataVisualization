#### 人邮模拟登录

```python
import requests
import matplotlib.pyplot as plt
url = "https://www.ptpress.com.cn/login"
s = requests.Session()
img = s.get("https://www.ptpress.com.cn/kaptcha.jpg?v=0.46106048366388075")
with open("./verifyCode.jpg", "wb") as f:
    f.write(img.content)
pic = plt.imread("./verifyCode.jpg")
plt.imshow(pic)  # 将pic可以用plt来进行显示
plt.show()
a = input("请手动输入验证码:")
login = {"username": "15976318994", "password": "15976318994", "verifyCode": a}
rqq = s.post(url, data = login)
print(rqq.url)
```
